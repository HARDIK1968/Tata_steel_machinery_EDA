**Simplified Explanation of Machine Failures with some Examples for better Understanding**

---


**Tool Wear Failure ** The machine tool becomes dull or breaks after
being used for a certain period. This happens randomly within a set usage range.

 **Example:** A drill bit wears out after many uses and eventually breaks.


**Heat Dissipation Failure ** If the machine doesnt cool down properly (due to low temperature difference between the air and the process) and runs at a slow speed, it can overheat and fail.

 **Example**: A laptop overheats because its cooling fan is blocked, causing it to shut down.


**Power Failure ** If the machine doesnt get enough power to run properly , it can stop working.

 **Example:** A car engine stalls if it doesnt get enough fuel to generate power.


**Overstrain Failure ** If the tool is too worn out and the machine is under too much stress , it may exceed safe limits and fail.

 **Example:** A machine part is heavily loaded beyond its capacity, causing it to crack or break.


**Random Failures ** Unexpected failures that happen by chance with a low probability, without any clear reason.

 **Example**: A perfectly working machine suddenly stops due to an unexpected internal defect.

Now we will Visualize with help of Barchart as to how many unique machines are availaible with tata stell and under which Quality they Fall.

```python
product_id_counts = combined_df.groupby('Type')['Product ID'].nunique().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=product_id_counts.index, y=product_id_counts.values ,palette="magma")
plt.xlabel("Type" , fontsize = 15)
plt.ylabel("Number of Unique Product IDs")
plt.title("Number of Unique Product IDs per Type")
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(product_id_counts.values):
    plt.text(i, v + 1, str(v), ha='center', va='bottom',fontsize=10 , fontweight = 'bold')

plt.tight_layout()
plt.show()
```

Now we will visualize how many failures are experienced by Different machine types with help of Barchart.

```python
machine_failures = train_df.groupby('Type')['Machine failure'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=machine_failures.index, y=machine_failures.values , palette="magma")
plt.xlabel("Machine Type" , fontsize = 15)
plt.ylabel("Number of Failures")
plt.title("Number of Failures per Machine Type")
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(machine_failures.values):
    plt.text(i, v + 1, str(v), ha='center', va='bottom',fontsize=10 , fontweight = 'bold')

plt.show()
```

So as we can see Low Quality machines face most failures, which is obvious given Most machines fall under this Category.

So now we will see the Percentage of Machine failures Compared to its Total Count , i.e How many Low , Medium and High Quality Machines have faced Failures in %.

```python
from tabulate import tabulate

total_machines = train_df.groupby('Type')['Machine failure'].count()
machine_failures = train_df.groupby('Type')['Machine failure'].sum()
failure_percentage = (machine_failures / total_machines) * 100

table_data = []
for machine_type in failure_percentage.index:
  table_data.append([machine_type, f"{failure_percentage[machine_type]:.2f}%"])

headers = ["Machine Type", "Failure Percentage"]
print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

So , as we can see from above Code result , inspite of Being Low in Quality , the Machines taht fall under the Low Quality have faced just 1.67% failures compared to High Quality machines with 1.30% failure.

So this tells us that Tata Steel can invest in more so-Called Low quality Machines as they are not in strict sense Low Quality and save On Capital Costs which would decrease their Capital expenditure (Machinery being a Capital asset) and in crease their overall Profitibility.

Now lets create A piechart to analyse and visualize which Failure type Contributes to Overall Failure type.

```python
failure_types = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']
machine_failures = combined_df.groupby('Product ID')[failure_types].sum().astype(bool).sum()

plt.figure(figsize=(6, 6))
plt.pie(machine_failures, labels=failure_types, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Failure Types Across Machines')
plt.axis('equal')
plt.show()
```

As we can see , OSF and HDF failure type dominate the Chart with 25.8% and 25.2% respectively. TWF failure type is lest common with just 10% count.
But overall , no particular failure type is dominant is Dataset.

Now lets analyse and visualize the how many Machines have faced how many Failures.

```python
failure_cols = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']
combined_df['Total_Failures'] = combined_df[failure_cols].sum(axis=1)

combined_df['Failure Category'] = combined_df['Total_Failures'].apply(lambda x:
    'No Failure' if x == 0 else
    'Single Failure' if x == 1 else
    '2 Failures' if x == 2 else
    '3 Failures' if x == 3 else
    '4 Failures' if x == 4 else
    '5 Failures')

unique_machines = combined_df[['id', 'Failure Category']].drop_duplicates()

failure_counts = unique_machines['Failure Category'].value_counts().reset_index()
failure_counts.columns = ['Failure Category', 'Unique Machines']

total_machines = failure_counts['Unique Machines'].sum()
failure_counts['Percentage'] = (failure_counts['Unique Machines'] / total_machines) * 100

category_order = ['No Failure', 'Single Failure', '2 Failures', '3 Failures', '4 Failures', '5 Failures']
failure_counts = failure_counts.set_index('Failure Category').reindex(category_order).reset_index()

plt.figure(figsize=(8,5))
ax = sns.barplot(x='Failure Category', y='Unique Machines', data=failure_counts, palette=['green', 'orange', 'red', 'blue', 'purple', 'brown'])

for p, percentage in zip(ax.patches, failure_counts['Percentage']):
    ax.text(p.get_x() + p.get_width() / 2,
            p.get_height() + 5,
            f'{int(p.get_height())} ({percentage:.1f}%)',
            ha='center', va='bottom', fontsize=8, fontweight='bold', color='black')

plt.title("Number of Machines by Failure Type", fontsize=14)
plt.xlabel("Failure Category", fontsize=12)
plt.ylabel("Unique Machine Count", fontsize=12)
plt.show()
```

So as we can see , 98.6% of the times , machines didn't face any failures. That means that Tata Steel's Production line is Well Maintained with regular Maintainence of Machineries for Optimal and Continuos flow of Production without any Breakdown.
